```python
true_values = []
for v in test_data_df.values:
    binary = v[8:]
    index = [k for k, i in enumerate(binary) if i == 1]
    for k in test_data_df.columns[8:][index]:
        true_values.append(v[0] + ' X ' + k)
```

```python
#Store Products ID in Multidimensional array
products =train_data_df[['P5DA', 'RIBP', '8NN1',
       '7POT', '66FJ', 'GYSR', 'SOP4', 'RVSZ', 'PYUQ', 'LJR9', 'N2MW', 'AHXO',
       'BSTQ', 'FM3X', 'K6QO', 'QBOL', 'JWFN', 'JZ9D', 'J9JW', 'GHYX', 'ECY3']]
products_t =test_data_df[['P5DA', 'RIBP', '8NN1',
       '7POT', '66FJ', 'GYSR', 'SOP4', 'RVSZ', 'PYUQ', 'LJR9', 'N2MW', 'AHXO',
       'BSTQ', 'FM3X', 'K6QO', 'QBOL', 'JWFN', 'JZ9D', 'J9JW', 'GHYX', 'ECY3']]

#Axis Transformation
train_data_df = train_data_df.melt(id_vars=train_data_df.columns[:8], value_vars=products, var_name = "PCODE", value_name="Label" )
test_data_df = test_data_df.melt(id_vars=test_data_df.columns[:8], value_vars=products, var_name = "PCODE", value_name="Label" )

#A merger
train_data_df['combiner']='x'

#Set
data=pd.concat([train_data_df,test_data_df],sort=False).reset_index(drop=True)
```

```python
data.head()
```

```python
#making our train data in the correct format
train_data_df=data[data.combiner.notnull()].reset_index(drop=True)
test_data_df=data[data.combiner.isna()].reset_index(drop=True)

#discarding the combiner column
train_data_df.drop('combiner', inplace=True, axis=1)

#Removing the target column from our test data
test_data_df.drop(['Label','combiner'], inplace=True, axis=1)
```

```python
#Concatenate new column towards submission requirement
train_data_df['ID X PCODE'] = train_data_df['ID'] + ' X ' + train_data_df['PCODE']
test_data_df['ID X PCODE'] = test_data_df['ID'] + ' X ' + test_data_df['PCODE']
```

```python
#convert the join_date to date time format
train_data_df["join_date"] = pd.to_datetime(train_data_df["join_date"])
test_data_df["join_date"] =pd.to_datetime(test_data_df["join_date"])
```

```python
#Separating join date 
train_data_df['join_day'] = train_data_df["join_date"].map(lambda x: x.day)
train_data_df['join_month'] = train_data_df["join_date"].map(lambda x: x.month)
train_data_df['join_year'] = train_data_df["join_date"].map(lambda x: x.year)
train_data_df['join_weekday'] = train_data_df['join_date'].dt.dayofweek
train_data_df['Age'] = train_data_df['join_year'] - train_data_df['birth_year']
train_data_df.drop('join_date',axis=1,inplace=True)
```

```python
#Separating join date 
test_data_df['join_day'] = test_data_df["join_date"].map(lambda x: x.day)
test_data_df['join_month'] = test_data_df["join_date"].map(lambda x: x.month)
test_data_df['join_year'] = test_data_df["join_date"].map(lambda x: x.year)
test_data_df['join_weekday'] = test_data_df['join_date'].dt.dayofweek
test_data_df['Age'] = test_data_df['join_year'] - test_data_df['birth_year']
test_data_df.drop('join_date',axis=1,inplace=True)
```

```python
#encoding our categorical data
from sklearn.preprocessing import OneHotEncoder
# Create the encoder.
ohe = OneHotEncoder(handle_unknown="ignore")
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train_data_df['sex'] = le.fit_transform(train_data_df['sex'])
train_data_df['marital_status'] = le.fit_transform(train_data_df['marital_status'])
train_data_df['occupation_code'] = le.fit_transform(train_data_df['occupation_code'])
train_data_df['occupation_category_code'] = le.fit_transform(train_data_df['occupation_category_code'])
train_data_df['PCODE'] = le.fit_transform(train_data_df['PCODE'])
train_data_df['branch_code'] = le.fit_transform(train_data_df['branch_code'])

test_data_df['sex'] = le.fit_transform(test_data_df['sex'])
test_data_df['marital_status'] = le.fit_transform(test_data_df['marital_status'])
test_data_df['branch_code'] = le.fit_transform(test_data_df['branch_code'])
test_data_df['occupation_code'] = le.fit_transform(test_data_df['occupation_code'])
test_data_df['occupation_category_code'] = le.fit_transform(test_data_df['occupation_category_code'])
test_data_df['PCODE'] = le.fit_transform(test_data_df['PCODE'])
```

```python
proba = xgb.predict_proba(test_data_df.drop(columns=['ID','ID X PCODE'], axis=1))[:, 1]
df_answer = test_data_df[['ID X PCODE']]
df_answer['Label'] = proba
```
